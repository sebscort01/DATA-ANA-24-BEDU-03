```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv(r"E:\Mt y Dip\Bedu\Modulo 3 Analisis de datos con python\DATA-ANA-24-BEDU-02\Ejemplo_04_Salary_Professions.csv")
data.head()
```

```python
# Convertir columna Salary a numerico
data['SALARY'] = pd.to_numeric(data['SALARY'], errors='coerce')

# Eliminar NaN
data = data.dropna(subset=['SALARY'])

data.head()

```

```python
# Crear un grafico
plt.figure(figsize=(10, 6))
sns.kdeplot(data['SALARY'], fill=True, color='b')
plt.title('Distribucion de Densidad')
plt.xlabel('Salario')
plt.ylabel('Densidad')
plt.grid(True)
plt.show()
```
